```python
import pandas as pd

df = pd.read_csv("~/Downloads/medicaid_parameters.csv")
df.head()
```

```python
category_to_openfisca_cat = {
    "0-1": "CHILD_AGE_0",
    "1-5": "CHILD_AGE_1_5",
    "6-18": "CHILD_AGE_6_18",
    "19-20": "ADULT_AGE_19_20",
    "PARENT": "ADULT_WITH_DEPENDENT",
    "19-64": "ADULT_WITHOUT_DEPENDENT",
    "PREGNANT": "PREGNANT",
}

for index, row in df.iterrows():
    print(f"{row.STATE}:")
    for cat, cat_key in category_to_openfisca_cat.items():
        if row[cat] == "-":
            value = "-.inf"
        else:
            value = float(row[cat].replace("%", "")) / 100
            value = f"{value:.2f}"
        print(f"  {cat_key}: \n    2021-01-01: {value}")
```
